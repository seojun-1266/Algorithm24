## 선택정렬 문제

반복문을 사용하여 저장된 값을 비교하고 가장 작은 값부터 출력하는 알고리즘

###알고리즘 개요 : 
함수 이름 : selection_sort
입력값 : data = [5, 6, 7, 9, 3, 5]
출력값 : data = [3, 5, 5, 6, 7, 9]

```python
#### 알고리즘 코드
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i
        for j in range(i + 1, n):
            if A[j] < A[least]:
                least = j
        A[i], A[least] = A[least], A[i]
      
  
data = [5, 6, 7, 9, 3, 5]  
print("Original: ", data)
selection_sort(data)
print("Selection: ", data)
```

```python
###테스트코드
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i
        for j in range(i + 1, n):
            if A[j] < A[least]:
                least = j
        A[i], A[least] = A[least], A[i]
      

data = [7, 8, 9, 5, 6, 3] 
print("Original: ", data)
selection_sort(data)
print("Selection: ", data)
```
