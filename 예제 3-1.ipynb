# 문제정의

제자리 정렬로 개선된 선택정렬 알고리즘

#알고리즘 설명
반복문을 사용하여 저장된 값을 비교하고 가장 작은 값부터 출력하는 알고리즘

# 알고리즘 개요 

함수 이름 : selection_sort
입력값 : data = [5, 6, 7, 9, 3, 5]
출력값 : data = [3, 5, 5, 6, 7, 9]
def selection_sort(A):                  #선택정렬
    n = len(A)                          #n:리스트  A의 길이
    for i in range(n-1):                #2중루프중 외부루프
        least = i                       
        for j in range(i + 1, n):       #2중루프중 내부루프
            if A[j] < A[least]:         #비교연산
                least = j               #최소항목 갱신
        A[i], A[least] = A[least], A[i] #배열 항목 교환
      

```python
# 알고리즘 코드
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i
        for j in range(i + 1, n):
            if A[j] < A[least]:
                least = j
        A[i], A[least] = A[least], A[i]
      
```

```python
# 테스트코드
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i
        for j in range(i + 1, n):
            if A[j] < A[least]:
                least = j
        A[i], A[least] = A[least], A[i]
      

data = [7, 8, 9, 5, 6, 3] 
print("Original: ", data)
selection_sort(data)
print("Selection: ", data)
```
