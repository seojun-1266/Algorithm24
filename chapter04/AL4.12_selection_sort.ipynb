# 문제정의
리스트 분할(Hoare Partition)


# 알고리즘 설명
리스트를 양쪽부터 하나씩 비교하며 조건에 맞게 바꾸거나 수정을 하며 서로 만날때까지 반복한다


# 코드 개요
(코드 개요 설명 1. 입력 변수 2. 출력 3. 함수설명)
입력 변수 : partition
출력 : 호출문으로써 메인 함수에서 나오는 값을 입력 받아 작동한다.
함수 설명 : 리스트의 양쪽 끝에서 부터 서로가 만날때까지 반복한다. 만약 조건에 해당된다면 서로 위치를 교환해주거나 수정을 해준다.


# 손으로 푼 예제 이미지 사진


```python
# 코드
def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) : # 왼쪽의 작은수가 오른쪽의 큰수보다 작으면 계속 실행
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :  # 왼쪽과 오른쪽이 만나서 교차하게되면 서로 교환을 해준다
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]  # 처음에 고정된 피벗을 high 위치와 교환한다
    return high                          # 피벗의 위치로 반환
```

```python
# 4.10코드
def kth_smallest_sort(A, k):
    A.sort()
    return A[k-1]

# 4.11코드
def quick_select(A, left, right, k):
    pos = partition(A, left, right)
    
    if (pos+ 1 == left+k):
        return A[pos]
    elif(pos+1 > left+k):
        return quick_select(A, left, pos-1, k)
    else :
        return quick_select(A, pos+1, right, k-(pos+1-left))

# 4.12코드
def partition(A, left, right) :
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) : # 왼쪽의 작은수가 오른쪽의 큰수보다 작으면 계속 실행
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high]> pivot : high-= 1
        
        if low < high :  # 왼쪽과 오른쪽이 만나서 교차하게되면 서로 교환을 해준다
            A[low], A[high] = A[high], A[low]
            
    A[left], A[high] = A[high], A[left]  # 처음에 고정된 피벗을 high 위치와 교환한다
    return high                          # 피벗의 위치로 반환

#실행코드(main)
array = [12, 3, 5, 7, 4, 19, 26, 23, 15]
print("입력 리스트= ",array)
print("[정렬기법] 3번째 작은 수:", kth_smallest_sort(array, 3))
print("[정렬기법] 6번째 작은 수:", kth_smallest_sort(array, 6))
n = len(array)
print("[축소기법] 3번째 작은 수:", quick_select(array, 0, n-1, 3))
print("[축소기법] 6번째 작은 수:", quick_select(array, 0, n-1, 6))
```
