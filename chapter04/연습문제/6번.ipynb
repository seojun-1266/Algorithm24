### 6. 셀 정렬(shell sort)은 삽입 정렬이 어느 정도 정렬된 배열에 대해서는 대단히 빠른 것에 착안한 정렬 방법이다. 이 정렬 방법을 찾아보고 삽입 정렬과의 차이를 설명해 보라.

셸 정렬(Shell Sort)은 도널드 셸(Donald Shell)에 의해 제안된 정렬 알고리즘으로, 삽입 정렬을 개선했다. 삽입 정렬과의 차이점은 이렇다.
<br/>
간격 설정: 삽입 정렬은 인접한 요소들을 비교하여 정렬하는 반면, 셸 정렬은 배열을 일정한 간격으로 나누어 각 부분 배열을 정렬한다. 간격은 초기에 큰 간격에서 시작하여 줄여나가는 방식으로 결정됩니다.
부분 배열 정렬: 삽입 정렬은 주어진 배열의 처음부터 끝까지 순차적으로 정렬하는 반면, 셸 정렬은 부분 배열을 정렬한다. 이러한 부분 배열 정렬은 배열의 일부가 이미 정렬된 경우에도 동작한다.
성능: 삽입 정렬은 배열이 거의 정렬되면 효율적이지만, 배열이 무작위로 섞여 있는 경우에는 성능이 떨어질 수 있다. 반면에 셸 정렬은 초기에 큰 간격으로 부분 배열을 정렬하여 전체 배열을 부분적으로 정렬하므로, 전체적으로 빠른 정렬이 된다. 삽입 정렬보다 평균적인 성능이 개선된다.
셸 정렬은 삽입 정렬의 성능을 개선하기 위해 제안된 알고리즘으로, 부분 배열의 정렬을 통해 배열의 거의 정렬된 부분을 활용하여 더욱 효율적으로 정렬을 수행한다.
