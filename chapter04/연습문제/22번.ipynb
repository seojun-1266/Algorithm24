### 22* 동일하게 생긴 12개의 동전과 양팔 저울이 있다. 동전 중에서 하나가 위조 동전이고, 정상동전과 무게가 약간 다른데, 무거운지 가벼운지는 알 수 없다. 양팔 저울을 세 번만 사용하여 어떤 동전이 위조 동전이고, 정상보다 무거운지 가벼운지를 판단하는 방법을 설계하라.
### (힌트: 동전은 세 그룹으로 나누어 비교해야 한다. 이것은 정말 어려운 문제이다.)

1번부터 12번까지의 동전들을 3개의 A, B, C 그룹으로 나눈다.
A = [1, 2, 3, 4], B = [5, 6, 7, 8], C = [9, 10, 11, 12]

1. 첫 번째 저울 사용을 해서 A그룹과 B그룹을 저울에 올려보자
2. 저울의 기울기는 변화가 없을것이다 A그룹과 B그룹에는 위조 동전이 없다는게 확인이 된다.
3. A그룹에서 3개의 동전을 가지고 C그룹의 3개의 임의의 동전을 가지고 양팔저울에 올렸을때,
양팔 저울에 움직임이 없다면 남은 한개가 위조 동전이다(ㄱ)
양팔저울에 움직임이 있다면 양팔 저울에 올려둔 동전 3개 중에 위조 동전이 있다.(ㄴ)

(ㄱ) C그룹에 있는 아무 동전 한개와 남아있는 동전 한개를 비교해서 위조 동전이 올라간다면 가벼운것이고 내려간다면 무거운것이다.
(ㄴ) A그룹과 비교했을때 내려갔다면 위조 동전은 무거운 것이고 위로 올라갔다면 가벼운것이다.
C그룹에 아무 동전 두 개를 비교해서 안움직인다면 나머지 한개가 위조 동전이고 변화가 있었다면 A그룹과 비교했을때의 무게와 같은 방향으로 가는쪽이 위조 동전이다.

1. 이전과 똑같이 A그룹과 B그룹을 저울에 올려보자
2. A그룹과 B그룹이 기울어 진다면 둘중 하나에 위조 동전이 있을것이다.(갈림길)
3. A그룹에서 동전 3개를 빼고, B그룹에서 동전 3개를 빼고 A그룹으로 옮기고, C그룹(정상동전그룹)에서 3개를 빼서 B그룹에 옮긴다.
4. 만약 양팔저울이 움직이지 않았다면 A그룹과 B그룹에 있던 한 개씩 남겨뒀던 동전들 중 위조 동전이 있다는 것을 알 수 있다.
5. A그룹에 있던 하나 남은 동전과 C그룹 동전을 양팔 저울에 올렸을때 양팔 저울이 수평을 이룬다면 B그룹에 남아 있던 동전이 위조 동전이며 가벼운 동전이 된다.

1. 이전과 똑같이 A그룹과 B그룹을 저울에 올려보자
2. A그룹과 B그룹이 기울어 진다면 둘중 하나에 위조 동전이 있을것이다.(갈림길)
3. A그룹에서 동전 3개를 빼고, B그룹에서 동전 3개를 빼고 A그룹으로 옮기고, C그룹(정상동전그룹)에서 3개를 빼서 B그룹에 옮긴다.
4. A그룹이 만약 내려가 있다가 올라왔다면 B그룹에서 가져온 3개중에 위조 동전이 있고 가벼운것을 확인 할 수 있다.
5. B그룹 에서 가져온 3개의 동전 중 2개를 올려 올라간 쪽이 위조 동전이고, 만약 수평을 이룬다면 나머지 하나가 위조 동전이다.
