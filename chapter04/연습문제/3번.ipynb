### 3. 다음 리스트를 삽입 정렬을 이용해 오름차순으로 정렬하라. 각 단계에서의 리스트의 내용을 적어라  

주어진 리스트는 [7 4 9 6 3 8 7 5]이다.
이를 오름차순으로 정렬하면 먼저 처음에는 첫 번째 항목 7을 이미 정렬된 부분 리스트로 간주한다.
그러므로 현재 리스트는 [7]이고, 나머지 부분은 [4, 9, 6, 3, 8, 7, 5] 이다.
다음으로 두 번째 항목 4를 적절한 위치에 삽입한다.
4보다 큰 숫자인 7 앞에 삽입하여 정렬한다.
그러므로 현재 리스트는 [4, 7]이고, 나머지 부분은 [9, 6, 3, 8, 7, 5]이다.
이와 같은 과정을 반복하여 리스트를 정렬힌다.

1. [7], [4, 9, 6, 3, 8, 7, 5]
2. [4, 7], [9, 6, 3, 8, 7, 5]
3. [4, 7, 9], [6, 3, 8, 7, 5]
4. [4, 6, 7, 9], [3, 8, 7, 5]
5. [3, 4, 6, 7, 9], [8, 7, 5]
6. [3, 4, 6, 7, 8, 9], [7, 5]
7. [3, 4, 6, 7, 7, 8, 9], [5]
8. [3, 4, 5, 6, 7, 7, 8, 9]
