# 문제정의
축소 정복 기법을 통해 팩토리얼을 계산한다.

# 알고리즘 설명
상향식 접근의 축소 정보 기법과 순환구조를 통한 팩토리얼의 계산


# 손으로 푼 예제 이미지 사진


# 코드 개요
입력변수 : n

출력 : n!

함수 설명 : 1부터 n+1까지의 범위를 설정한 뒤, 결과값 1에 n+1값이 범위의 끝에 도달할 때 까지 연산을 반복한다.


```python
# 코드
def factorial_iter(n):
    result=1
    for k in range(1,n+1):
        result=result*k
    return result
```

```python
# 테스트 코드
def factorial_iter(n):
    result=1
    for k in range(1,n+1):
        result=result*k
    return result
n=int(input("정수 n 값을 입력하시오: "))
factorial_iter(n)
print("%d! = %d",n,factorial_iter(n))
 
```

# 복잡도 분석
변수 n의 값만큼 계산하므로 O(n)이다.
