# 문제정의
이진트리의 후위 순회 문제 

# 알고리즘 설명
왼쪽부터 오른쪽으로 가는데 중위 순회와 달리 최대한 아래왼쪽에서부터 올라가며 오른쪽 노드 기준 첫 맨 아래 왼쪽부터 계산된다.

# 코드 개요
입력 : n, 트리
출력 : 호출문으로 호출을 할 시 가장 왼쪽 서브 트리부터 처리를 하여 순차적으로 오른쪽 서브 트리까지 처리한다.
함수내용 : 
postorder(n) : 후위순회 하는 함수


```python
# 코드
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def postorder(n):
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')
        

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
postorder(root)
print()
        
```

```python
# 테스트 코드
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def postorder(n):
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')

z = TNode('Z', None, None)
i = TNode('I', None, z)
h = TNode('H', None, None)
p = TNode('P', None, None)
g = TNode('G', None, p)
d = TNode('D', g, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', h, i)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(' In-Order : ', end=' ')
postorder(root)
print()
```
