# 문제정의
퀵정렬은 평균적으로 매우 빠른 수행 속도를 자랑하는 알고리즘으로 역시 분할정복 알고리즘을 사용한다. 그러나 병합 정렬이 항목들을 위치에 따라 분할하는 것과는 달리 퀵 정렬은 항목의 값에 따라 분할한다.

# 코드 개요
입력 변수 : A, left, right
출력 : 정렬된 A
코드 개요 : 

def quick_sort(A, left, right):         #A[left..right]를 오름차순
    if left < right:                    #정렬 범위가 2개 이상인 경우
        mid = partition(A, left, right) #좌우로 분할
        quick_sort(A, left, mid - 1)    #왼쪽 부분리스트를 퀵 정렬
        quick_sort(A, mid + 1, right)   #오른쪽 부분 리스트를 퀵정렬





```python
# 코드
def partition(A, left, right):
    pivot = A[right]
    i = left - 1
    for j in range(left, right):
        if A[j] <= pivot:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[right] = A[right], A[i + 1]
    return i + 1

def quick_sort(A, left, right):
    if left < right:
        mid = partition(A, left, right)
        quick_sort(A, left, mid - 1)
        quick_sort(A, mid + 1, right)

```

```python
# 테스트 코드

def partition(A, left, right):
    pivot = A[right]
    i = left - 1
    for j in range(left, right):
        if A[j] <= pivot:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[right] = A[right], A[i + 1]
    return i + 1

def quick_sort(A, left, right):
    if left < right:
        mid = partition(A, left, right)
        quick_sort(A, left, mid - 1)
        quick_sort(A, mid + 1, right)

data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original:", data)
quick_sort(data, 0, len(data) - 1)
print("QuickSort:", data)
```
