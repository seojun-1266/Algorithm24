# 문제정의
최근접 쌍의 거리 문제를 억지기법이 아닌 분할 정복 기법으로 만든 알고리즘으로 사용한다.

# 알고리즘 설명
분할 정복 기법을 기반으로 하며 각 부분에서 최근접 쌍을 찾은 다음 부분해를 통합하여 전체적인 최근접 쌍을 찾는 방법이다

# 코드 개요
입력 : P = [(2,6),(4,7),(4,1),(5,1),(3,2),(6,5),(7,9),(3,4),(0,1)]  
출력 : 정렬된 P = [(0, 1), (4, 1), (5, 1), (3, 2), (3, 4), (2, 6), (6, 5), (4, 7), (7, 9)]  
함수 내용 :
strip_closest(P, d): P와 d를 입력받아서 d보다 작은 최근저 쌍의 거리를 계산하는 함수


```python
# 코드
import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2) #3장 113p

def strip_closest(P, d):
    n = len(P)
    d_min = d

    for i in range(n):
        j = i + 1
        while j < n and P[j][1] - P[i][1] < d_min: 
            dij = distance(P[i], P[j])
            if dij < d_min:
                d_min = dij
            j += 1
    return d_min
```

```python
# 테스트 코드

import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2) #3장 113p

def strip_closest(P, d):
    n = len(P)
    d_min = d

    for i in range(n):
        j = i + 1
        while j < n and P[j][1] - P[i][1] < d_min: 
            dij = distance(P[i], P[j])
            if dij < d_min:
                d_min = dij
            j += 1
    return d_min
P = [(2,6),(4,7),(4,1),(5,1),(3,2),(6,5),(7,9),(3,4),(0,1)]
P.sort(key=lambda point: point[1])  
print("영역내에서 d보다 작은 최근접쌍의 거리:", strip_closest(P, len(P)))
```
