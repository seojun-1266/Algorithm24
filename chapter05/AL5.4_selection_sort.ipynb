# 문제정의
제귀적인 방법으로 트리의 높이를 구하는 문제

# 알고리즘 설명
양쪽 트리가 있는지 확인하고 누가 더 높은지 확인하여 더 높은 곳으로 이동하게 되고 이 이동 횟수를 트리의 높이로 반환함

# 코드 개요
입력 : root
출력 : 해당 트리의 높이
함수 내용 :   
calc_heigh() 트리의 높이를 알려주는 함수

```python
# 알고리즘 코드
def calc_height(root):
    if root is None:
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft, hRight) + 1
        
```

```python
# 테스트 코드
class TreeNode:
    def __init__(self, key, left=None, right=None):
        self.key = key
        self.left = left
        self.right = right

def calc_height(root):
    if root is None:
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft, hRight) + 1

d = TreeNode('D')
e = TreeNode('E')
b = TreeNode('B', d, e)
f = TreeNode('F')
c = TreeNode('C', f)
root = TreeNode('A', b, c)

print("트리의 높이:", calc_height(root))
```
