# 문제정의
 허프만 트리 생성 알고리즘

# 알고리즘 설명
주어진 빈도와 레이블로 최소 힙 초기화 후에 힙에서 두개의 최소 빈도 요소를 꺼내 합쳐 새로운 노드를 만들고 힙에 다시 삽입 


# 코드 개요
입력 : freq, label
출력 : 최소 빈도 노드를 더하는 과정, 생성된 허프만 트리


# 손으로 푼 예제 이미지 사진 


```python
# 코드
import heapq
def make_heap_tree(freq, label):
    n = len(freq)
    h = []
    for i in range(n):
        heapq.heappush(h,(freq[i], label[i]))
    
    for i in range(1, n):
        e1 = heapq.heappop(h)
        e2 = heapq.heappop(h)
        heapq.heappush(h, (e1[0]+e2[0], e1[1]+e2[1]))
        print(e1, "+", e2)
    print(heapq.heappop(h))
     
```
