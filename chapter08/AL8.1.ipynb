# 문제정의
거스름돈 동전 최소화 알고리즘

# 알고리즘 설명
거스름돈의 액수가 주어지면 액면가가 가장 높은 동전을 이용하여 먼저 잔돈의 액수를 최대한으로 줄인다. 남은 액수는 다음 액면가의 동전을 최대한으로 사용하는 과정을 순서대로 반복한다.


# 손으로 푼 예제 이미지 사진 



# 코드 개요
함수명: min_coins_greedy(coins, V)
입력 변수: coins, V
출력: count

```python
# 코드
def min_coins_greedy(coins, V):
    count = []
    remaining_value = V  # V를 새로운 변수로 복사하여 사용
    for i in range(len(coins)):
        cnt = remaining_value // coins[i]
        count.append(cnt)
        remaining_value -= cnt * coins[i]
    return count
```

```python
# 테스트 코드

def min_coins_greedy(coins, V):
    count = []
    remaining_value = V  # V를 새로운 변수로 복사하여 사용
    for i in range(len(coins)):
        cnt = remaining_value // coins[i]
        count.append(cnt)
        remaining_value -= cnt * coins[i]
    return count

coins = [500, 100, 50, 10, 5, 1]
V = 580
print("잔돈 액수 = ", V)
print("동전 종류 = ", coins)
print("동전 개수 = ", min_coins_greedy(coins, V))
```
