문제 정의 : 분할 가능한 배낭 채우기(탐욕적 기법)

# 알고리즘 설명

1. 탐욕적 기법을 사용하여 해결할 수 있는 대표적인 최적화 문제다. 
2. 가치 댑; 무게 비율이 높은 물건부터 배낭에 담는 방식으로 문제 해결한다.

# 코드 개요 :

1. 입력 변수 : items : 각 물건의 정보가 담긴 리스트. 각 항목은 튜플로 구성되며 (이름, 무게, 가치) 형태이다.
2. 출력 변수 : total_value : 배낭에 담을 수 있는 물건의 총 가치를 반환한다.
3. 함수 설명 : items = sorted(items, key=lambda x: x[2] / x[1], reverse=True) 각 물건의 가치 대비 무게 비율을 계산하여 내림차순으로 정렬한다.
return total_value : 배낭에 담긴 물건의 총가치를 반환한다.

# 손으로 푼 예제 :

# 알고리즘 코드

```python
def knapSack_fractions_greedy(odj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)

    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
            W<= o[1]
            totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))

    return totalValue
```

# 테스트 코드

```python
def knapSack_fractional_greedy(items, W):
    
    items = sorted(items, key=lambda x: x[2] / x[1], reverse=True)
    total_value = 0
    for item in items:
        if W >= item[1]:
            W -= item[1]
            total_value += item[2]
        else:
            total_value += item[2] * (W / item[1])
            break
    return total_value

obj1 = [("A", 10, 90), ("B", 20, 200), ("C", 8, 60)]
print("W = 18 ", obj1)
print("부분적인배낭(18): ", knapSack_fractional_greedy(obj1, 18), end='\n\n')

obj2 = [("A", 10, 100), ("B", 70, 70), ("C", 40, 160), ("D", 50, 250)]
print("W = 50 ", obj2)
print("부분적인배낭(50): ", knapSack_fractional_greedy(obj2, 50))
```
