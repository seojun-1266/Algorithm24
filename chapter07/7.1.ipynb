>##### (코드 개요 설명 1. 입력 변수 2. 출력 3. 함수설명)
>1. 입력 변수 : fib_dp_mem
>2. 출력      &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;: 피보나치(n) 을 출력함
>3. 함수 설명 : 피보나치를 계산하기 위해서 분할정복기법으로 구현한 하향식으로 문제를 해결한다. 한번 해결한 문제는 답을 메모리에 저장해 재활용함

---------

#### **5. 코드(PY, C)**

```python
def fib_dp_mem(n):
    if mem[n] is None:
        if n < 2:
            mem[n] = n
        else:
            mem[n] = fib_dp_mem(n-1) + fib_dp_mem(n-2)
    return mem[n]

n = 8
mem = [None] * (n+1)
print('동적계획(메모이제이션): Fibonacci(%d) = ' % n, fib_dp_mem(n))
```

------

#### **6. 테스트 코드(PY, C), 입력을 바꾸어서**

```python
def fib_dp_mem(n):
    if mem[n] is None:
        if n < 2:
            mem[n] = n
        else:
            mem[n] = fib_dp_mem(n-1) + fib_dp_mem(n-2)
    return mem[n]

n = 11
mem = [None] * (n+1)
print('동적계획(메모이제이션): Fibonacci(%d) = ' % n, fib_dp_mem(n))
```
