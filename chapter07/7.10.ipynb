# 문제정의
Floyd 최단경로 알고리즘

# 알고리즘 설명
이 알고리즘은 모든 정점 간의 최단 경로를 찾아 현재까지 알려진 최단 경로를 최신화 하여 최종적인 정점 간의 최단 경로를 찾는 알고리즘


# 손으로 푼 예제 이미지 사진


# 코드 개요
입력 : vertex, weight  
출력 : 모든 정점 간의 최단경로  
함수 설명 :  
 shortest_path_floyd() : 모든 정점 쌍 간의 최단경로를 계산하고 결과 출력
 

```python
# 코드

import copy
def shortest_path_floyd(vertex, W):
    vsize = len(vertex)
    D = copy.deepcopy(W)

    for k in range(vsize) :
        for i in range(vsize) :
            for j in range(vsize) :
                if (D[i][k] + D[k][j] < D[i][j]) :
                    D[i][j] = D[i][k] + D[k][j]
        printD(D)
```

```python
#테스트 코드
import copy
def shortest_path_floyd(vertex, W):
    vsize = len(vertex)
    D = copy.deepcopy(W)

    for k in range(vsize) :
        for i in range(vsize) :
            for j in range(vsize) :
                if (D[i][k] + D[k][j] < D[i][j]) :
                    D[i][j] = D[i][k] + D[k][j]
        printD(D)
def printD(D):
    vsize = len(D)
    print("=================================")
    for i in range(vsize) :
        for j in range(vsize) :
            if (D[i][j] == INF) : print(" INF ", end='')
            else : print("%4d "%D[i][j], end='')
        print("")
INF = 9999
vertex = [ 'A', 'B', 'C', 'D', 'E', 'F', 'G' ]
weight = [[   0,  7, INF, INF,   3, 10,  INF],
          [   7,  0,   4,  10,   2,  6,  INF],
          [ INF,  4,   0,   2, INF, INF, INF],
          [ INF, 10,   2,   0,  11,   9,   4],
          [   3,  2, INF,  11,   0,  13,   5],
          [  10,  6, INF,   9,  13,   0, INF],
          [ INF,INF, INF,   4,   5, INF,   0]]
print("Shortest Path By Floyd's Algorithm")
shortest_path_floyd(vertex, weight)
```
